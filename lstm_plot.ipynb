```python
import matplotlib.pyplot as plt
```

```python
def plot_metriks(data,title):
    names = list(data.keys())
    values = list(data.values())
    plt.title(title)
    plt.bar(range(len(data)), values, tick_label=names)
    plt.savefig(title, dpi=1200)
    plt.show()
```
